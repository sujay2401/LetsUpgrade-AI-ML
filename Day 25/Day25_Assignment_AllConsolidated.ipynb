```python
y = dataset["Sex"] # DV
```

```python
list=[]
list = dataset.drop(["PassengerId","Cabin","Ticket","Name"],axis=1)
print(list)
list
```

```python
def choose_column(label):
    print(label)
    X = dataset.drop([label,"PassengerId","Cabin","Ticket","Name"],axis=1)
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=0)
    bnl = BernoulliNB()
    y_pred = bnl.fit(X_train,y_train).predict(X_test)
    conf_matrix = confusion_matrix(y_test,y_pred)
    print(len(conf_matrix))
    print("\n..Accuracy_Score for label",label," is..\n")
    print(accuracy_score(y_test,y_pred,normalize=True))
    print("\n..Confusion Matrix for label",label," is..\n")
    print(confusion_matrix(y_test,y_pred))
    print("Total",np.sum(confusion_matrix(y_test,y_pred)))
    print("Correct",np.trace(confusion_matrix(y_test,y_pred)))
    print("=========End of ",label,"===============================")
    
for i in list.columns:
    print(i)
    choose_column(i)
```
