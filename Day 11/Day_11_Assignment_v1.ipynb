```python
import pandas as pd
from scipy.stats import pearsonr as pr



dataset1 = pd.read_excel('Attrition_general_data.xlsx',sheetname=0)

d = {'No': 0, 'Yes': 1}
dataset1['Attrition'] = dataset1['Attrition'].map(d).fillna(dataset1['Attrition'])
dataset1['Attrition'].round(0).astype(int)
dataset1
#print('r,p',pr(dataset1.Attrition,dataset1.DistanceFromHome))
```

```python
'''
1st Hypothesis
--------------
H0 = There is no signficant difference between Attrition and DistanceFromHome
H1 = There is signficant difference between Attrition and DistanceFromHome
'''
print('r,p',pr(dataset1.Attrition,dataset1.DistanceFromHome))

#Conclusion: Since Probability value is 0.51828 which means P is greater than 0.05. Therefore Null hypothesis is Accepted 
#and Alternate hypothesis is rejected.Also since r = -0.009730 corelation is negative or no correlation 
```
