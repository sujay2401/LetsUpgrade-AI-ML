```python
'''Appling Man Witney test
H0 = There is no significant difference between Age for both Attrition_yes and Attrition_no
H1 = There is significant difference between Age for both Attrition_yes and Attrition_no

'''
import pandas as pd

dataset_yes= pd.read_excel('Attrition_general_data.xlsx',sheetname="Attrition_yes")
dataset_no= pd.read_excel('Attrition_general_data.xlsx',sheetname="Attrition_No")

from scipy.stats import mannwhitneyu

stats,p = mannwhitneyu(dataset_yes.Age, dataset_no.Age)

print(stats,p)

# Conclusion - Since p = 2.9951588479067175e-30 Probability value is < 0.05 which mean we can say that There is significant difference 
#between Age for both Attrition_yes and Attrition_no (Alternative Hypothesis is accepted)
```

## Inferance of the above test with Attrition_yes and No vs Age
<io>
    <li>H0 = There is no significant difference between Age for both Attrition_yes and Attrition_no</li>
    <li>H1 = There is significant difference between Age for both Attrition_yes and Attrition_no</li>
</io>

### Here if we see that the probability value is 2.9951e-30 which is lesser than 0.05
### Therefore we can conclude that Alternative hypothesis is accepted i,e There is significant difference between Age for both Attrition_yes and Attrition_no
